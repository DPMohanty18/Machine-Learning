# Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])


figure(figsize=(4, 4), dpi=80)
plt.bar(x,y)
plt.show()
```

## Histogram

```python
import numpy as np

x = np.random.normal(170, 10, 250)
figure(figsize=(2,2))
plt.hist(x)
plt.show() 
```

## Pie chart

```python
y = np.array([35, 25, 25, 15])
figure(figsize=(2,3))
cars = ['AUDI', 'BMW', 'FORD','TESLA']
plt.pie(y,labels=cars)
plt.show() 
```

## box plot

```python
l=[i for i in range(0,100)]
figure(figsize=(2,3))
plt.boxplot(l)
plt.show() 
```

## Scatter plot

```python
x=np.random.normal(0,1,100)
y=np.random.normal(0,2,100)
figure(figsize=(3,3))
colors = np.random.randint(50, 150, size=(100))
plt.scatter(x,y,c=colors,cmap='plasma')
plt.show()
```
