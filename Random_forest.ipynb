```python
data=data.drop("Loan_ID",axis=1)
```

```python
traget=data["Loan_Status"]
features=data.drop("Loan_Status",axis=1)
```

```python
data2=pd.get_dummies(features,dtype='int')
```

```python
data2.head()
```

## Sampling

```python
target_data=traget.map({'N':0,'Y':1})
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data2.values,target_data.values,test_size=0.2)
```

## Model building

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

## Performance checking

```python
pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
